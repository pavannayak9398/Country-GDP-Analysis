```python
# Introduction to seaborn(Statistic Visulaization- random distribution)

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
plt.rcParams['figure.figsize']=6,2

import warnings
warnings.filterwarnings('ignore')
```

```python
df['InternetUsers']
```

```python
#Distplot stands for distribution plot,
#it takes as input an array and plots a curve corresponding to the distribution of points in the array.

sns.distplot([0, 1, 2, 3, 4, 5])
```

```python
#without histogram
sns.distplot([0, 1, 2, 3, 4, 5],hist=False)
```

```python
#distplot
sns.distplot(df['InternetUsers']) #Univariate analysis
```

```python
#displot
sns.displot(df['InternetUsers']) #Univariate analysis
```

```python
#box plot
sns.boxplot(data=df,x='IncomeGroup',y='BirthRate') #bivariate analysis
```

```python
#lmplot-linear model plot
sns.lmplot(data=df,x='InternetUsers',y='BirthRate')
```

```python
sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=False)
```

```python
sns.lmplot(data=df,x='InternetUsers',y='BirthRate',fit_reg=True)
```
